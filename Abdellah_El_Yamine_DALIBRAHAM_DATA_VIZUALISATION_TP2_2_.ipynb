```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Reading the data
df = pd.read_csv('/content/gym_members_exercise_tracking.csv')
```

## 1. Scatter Plot (e.g., Age vs Calories Burned)

```python
# Matplotlib Scatter Plot

plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Calories_Burned'], color='blue')
plt.title('Age vs Calories Burned (Matplotlib)')
plt.xlabel('Age')
plt.ylabel('Calories Burned')
plt.grid(True)
plt.show()
```

```python
# Seaborn Scatter Plot

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Calories_Burned', data=df, hue='Experience_Level', palette='coolwarm')
plt.title('Age vs Calories Burned by Experience Level (Seaborn)')
plt.xlabel('Age')
plt.ylabel('Calories Burned')
plt.grid(True)
plt.show()
```

```python
# Plotly Express Scatter Plot

fig = px.scatter(df, x='Age', y='Calories_Burned', color='Experience_Level',
                 title='Age vs Calories Burned by Experience Level (Plotly Express)',
                 labels={'Calories_Burned': 'Calories Burned'})
fig.show()
```

## 2. Box Plot (e.g., Calories Burned by Workout Type)

```python
# Matplotlib Box Plot

plt.figure(figsize=(8, 6))
df.boxplot(column='Calories_Burned', by='Workout_Type', patch_artist=True)
plt.title('Calories Burned by Workout Type (Matplotlib)')
plt.suptitle('')  # Removes automatic subtitle
plt.xlabel('Workout Type')
plt.ylabel('Calories Burned')
plt.xticks(rotation=45)
plt.show()
```

```python
# Seaborn Box Plot

plt.figure(figsize=(8, 6))
sns.boxplot(x='Workout_Type', y='Calories_Burned', data=df, palette='Set3')
plt.title('Calories Burned by Workout Type (Seaborn)')
plt.xlabel('Workout Type')
plt.ylabel('Calories Burned')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotly Express Box Plot

fig = px.box(df, x='Workout_Type', y='Calories_Burned', title='Calories Burned by Workout Type (Plotly Express)',
             labels={'Calories_Burned': 'Calories Burned'})
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

## 3. Bar Plot (e.g., Average Max BPM by Workout Type)

```python
# Calculate mean Max BPM by Workout Type
max_bpm_avg = df.groupby('Workout_Type')['Max_BPM'].mean()
```

```python
# Matplotlib Bar Plot

plt.figure(figsize=(8, 6))
plt.bar(max_bpm_avg.index, max_bpm_avg.values, color='green')
plt.title('Average Max BPM by Workout Type (Matplotlib)')
plt.xlabel('Workout Type')
plt.ylabel('Average Max BPM')
plt.xticks(rotation=45)
plt.show()
```

```python
# Seaborn Bar Plot

plt.figure(figsize=(8, 6))
sns.barplot(x='Workout_Type', y='Max_BPM', data=df, palette='viridis')
plt.title('Average Max BPM by Workout Type (Seaborn)')
plt.xlabel('Workout Type')
plt.ylabel('Average Max BPM')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotly Express Bar Plot

fig = px.bar(df, x='Workout_Type', y='Max_BPM', title='Average Max BPM by Workout Type (Plotly Express)',
             labels={'Max_BPM': 'Average Max BPM'}, color='Workout_Type')
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

## 4. Heatmap (Correlation Matrix between Numerical Variables)

```python
# Correlation matrix
corr = df[['Age', 'Weight (kg)', 'Height (m)', 'Max_BPM', 'Avg_BPM', 'Resting_BPM',
           'Session_Duration (hours)', 'Calories_Burned', 'Fat_Percentage',
           'Water_Intake (liters)', 'Workout_Frequency (days/week)', 'BMI']].corr()
```

```python
# Seaborn Heatmap


plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix (Seaborn)')
plt.show()
```

```python
# Plotly Express Heatmap

# Heatmap using Plotly Express
fig = px.imshow(corr, text_auto=True, aspect="auto", title="Correlation Matrix (Plotly Express)")
fig.show()
```
